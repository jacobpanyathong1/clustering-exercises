```python
def row_col(df):
    b = list(df.columns)
    a= df.isna().sum()
    d= pd.Series(a).values
    c= {'columns': b, 'num_rows_missing' : d, 'pct_rows_missing' : a/len(df)}
    df = pd.DataFrame(data=c)
    df.reset_index(inplace=True)
    return df
```

###### Create a function that will drop rows or columns based on the percent of values that are missing: handle_missing_values(df, prop_required_column, prop_required_row).

```python
def handle_missing_values(df, prop_required_column, prop_required_row):
    b = list(df.columns)
    a= df.isna().sum()
    d= pd.Series(a).values
    c= {'columns': b,'num_rows_missing' : d, 'pct_rows_missing' : a/len(df)}
    new_df = pd.DataFrame(data=c)
    new_df.reset_index(drop=True, inplace=True)
    
    cols = df.columns
    for col in cols:
        if new_df.pct_rows_missing >= prop_required_row & :
            df.dropna(thresh = "{:.0%}".format({prop_required_row}), inplace=True)
        if df.num_rows_missing >= prop_required_column:
            df.drop(columns=[col], inplace=True) 
            return df
    
    
```
