```python
def row_col(df):
    b = list(df.columns)
    a= df.isna().sum()
    d= pd.Series(a).values
    c= {'columns': b, 'num_rows_missing' : d, 'pct_rows_missing' : a/len(df)}
    df = pd.DataFrame(data=c)
    df.reset_index(inplace=True)
    return df
```

###### Create a function that will drop rows or columns based on the percent of values that are missing: handle_missing_values(df, prop_required_column, prop_required_row).

```python
def handle_missing_values(df, prop_required_column, prop_required_row):
    
    cols = df.columns
    prop_null_column = 1 - prop_required_column
    for col in cols:
        
        n_sum = df[col].isna().sum()
        n_pct = n_sum / len(df)
        
        if n_pct > prop_null_column:
            df.drop(columns=col, inplace=True)
            
    row_threshold = int(prop_required_row * df.shape[1])
    
    df.dropna(axis=0, thresh=row_threshold, inplace=True)
    
    return df
    
    
```

```python
def out_calc(df, variable):
    q1 = np.quantile(df[variable], .25)
    q3 = np.quantile(df[variable], .75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q1 + 1.5 * iqr
    
    return print(f'lower: {lower}, upper: {upper}')
```
